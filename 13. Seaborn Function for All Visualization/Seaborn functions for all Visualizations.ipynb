```python
#Loading Packages
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style='darkgrid')
```

```python
#Loading Data
df = pd.read_csv("Data/supermarket.csv", parse_dates=['Date'])
df.head()
```

# Relplot

```python
fig = sns.relplot(data=df, x='Unit price', y='Total', kind='scatter', height=5, aspect =2.5)
fig.savefig('Plots/Scatter.jpg')
```

```python
df_sub = df[['Total', 'Date']].groupby('Date').sum().reset_index()

df_sub.head()
```

```python
fig = sns.relplot(data=df_sub, x='Date', y='Total', kind='line', height=5, aspect=2.5)
fig.savefig('Plots/Line.jpg')
```

# Distribution Plot

```python
fig = sns.displot(data = df, x='Total', hue='Gender', kind='hist', multiple='dodge', palette='Blues',height=5, aspect=2.5)
fig.savefig('Plots/Hist.jpg')
```

```python
fig = sns.displot(data = df, x='Total', hue='Gender', kind='kde', palette='cool', height=5, aspect=2.5)
fig.savefig('Plots/KDE.jpg')
```

# Categorical Plot

```python
fig = sns.catplot(data=df, x='Product line',y='Unit price', kind='box', height=5, aspect=2.5, width=0.5) 
fig.savefig('Plots/Box.jpg')
```

```python
fig = sns.catplot(data=df, x='Branch', y='Total', kind='strip', height=5, aspect=2.7)
fig.savefig('Plots/Strip.jpg')
```

```python
fig = sns.catplot(data=df, x='Branch', y='Total', kind='violin', height=5, aspect=3)
fig.savefig('Plots/Violin.jpg')
```
