# The Heatmap with Time

```python
import datetime

ROBBERY['year'] = pd.DatetimeIndex(ROBBERY['ARREST_DATE']).year
ROBBERY['month'] = pd.DatetimeIndex(ROBBERY['ARREST_DATE']).month

temp_list = []
for x in ROBBERY['month']:
    
    monthinteger = x
    
    temp_list.append(datetime.date(1900, monthinteger, 1).strftime('%B'))
    
ROBBERY['month_in_full'] = temp_list
ROBBERY['month_year'] = [d.split('/')[2] + " " + d.split('/')[0] for d in ROBBERY.ARREST_DATE]
```

```python
#first we need to create weight list
weight_list = []
ROBBERY_09_19['conta'] = 1

#loop through the weight list
for x in ROBBERY_09_19['month_year'].sort_values().unique():
    weight_list.append(ROBBERY_09_19.loc[ROBBERY_09_19['month_year'] == x,
                                        ['Latitude', 'Longitude', 'conta']].groupby(['Latitude', 'Longitude']).sum().reset_index().values.tolist())
```

```python
#create the base map
my_base_map = folium.Map(location = [40.74, -73.98], control_scale = True, zoom_start=10, tiles = 'Stamen Toner')

#create the heatmap from our list
plugins.HeatMapWithTime(weight_list, radius=10, gradient={0.1:'blue', 0.25:'green', 0.5:'yellow', 0.75:'orange',1:'red'},
               auto_play=True, min_opacity=0.5, max_opacity=1, use_local_extrema=True).add_to(my_base_map)

my_base_map
```

```python
my_base_map.save("index.html")
```
